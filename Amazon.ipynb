```python
# top 3 completed payment methods
from sqlalchemy import text


query1 = text("""
    SELECT payment_method, status, COUNT(*) AS status_count
    FROM orders_no_id
    GROUP BY payment_method, status
    ORDER BY payment_method, status;
""")


query2 = text("""
    SELECT payment_method, COUNT(*) AS completed_count
    FROM orders_no_id
    WHERE status = 'Completed'
    GROUP BY payment_method
    ORDER BY completed_count DESC
    LIMIT 3;
""")

with engine.connect() as conn:
    result1 = conn.execute(query1)
    print("     Order Status Counts by Payment Method")
    for row in result1:
        print(row)

    result2 = conn.execute(query2)
    print("\n   Top 3 Completed Payment Methods ")
    for row in result2:
        print(row)
```

```python
from sqlalchemy import text

query = text("""
    SELECT
        payment_method,
        SUM(CASE WHEN status = 'Completed' THEN 1 ELSE 0 END) AS completed_count,
        SUM(CASE WHEN status = 'Pending' THEN 1 ELSE 0 END) AS pending_count,
        SUM(CASE WHEN status = 'Cancelled' THEN 1 ELSE 0 END) AS cancelled_count,
        COUNT(*) AS total_orders
    FROM orders_no_id
    GROUP BY payment_method
    ORDER BY completed_count DESC
    LIMIT 3;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("     Top 3 Payment Methods with Status Breakdown")
    for row in result:
        print(row)
```

```python
#Category has the most sales in Bostan
from sqlalchemy import text

query = text("""
    SELECT
        category,
        SUM(total_sales) AS total_sales_in_boston
    FROM orders_no_id
    WHERE customer_location = 'Boston'
    GROUP BY category
    ORDER BY total_sales_in_boston DESC;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("Top Category by Total Sales in Boston:")
    for row in result:
        print(row)
```

```python
# 2nd most costly product in HA (Sorting by price)
from sqlalchemy import text

query = text("""
    SELECT product, price
    FROM orders_no_id
    WHERE category = 'Home Appliances'
    ORDER BY price DESC
    LIMIT 1 OFFSET 1;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("Second Most Costly Product in Home Appliances")
    for row in result:
        print(row)
```

```python
# List of all unique Customers (Distinct)
from sqlalchemy import text

query = text("""
    SELECT DISTINCT customer_name, category
    FROM orders_no_id
    ORDER BY customer_name;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("List of Unique Customers")
    for row in result:
        print(row)
```

```python
# Name starting with 'E' ()
from sqlalchemy import text

query = text("""
    SELECT DISTINCT customer_name
    FROM orders_no_id
    WHERE customer_name LIKE '%vi%'
    ORDER BY customer_name;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("Customers Whose Names Start with 'E':")
    for row in result:
        print(row)
```

```python
#All orders placed in March
from sqlalchemy import text

query = text("""
    SELECT order_id, order_date
    FROM orders_no_id
    WHERE MONTH(order_date) = 3
    ORDER BY order_date;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("Orders Placed in March")
    for row in result:
        print(row)
```

```python
from sqlalchemy import text

query = text("""
    SELECT DISTINCT 
        a.customer_name AS customer1,
        b.customer_name AS customer2,
        a.customer_location
    FROM orders_no_id a
    JOIN orders_no_id b
        ON a.customer_location = b.customer_location
       AND a.customer_name < b.customer_name
    ORDER BY a.customer_location, customer1, customer2;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("Pairs of Customers from the Same Location")
    for row in result:
        print(row)
```

```python

# date has the highest sales
from sqlalchemy import text

query = text("""
    SELECT order_date, SUM(total_sales) AS total_sales
    FROM orders_no_id
    GROUP BY order_date
    ORDER BY total_sales DESC
    LIMIT 1;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("Date with the Highest Sales")
    for row in result:
        order_date = row[0]  
        total_sales = row[1] 
        print(f"Date: {order_date}, Total Sales: {total_sales}")
```

```python
# first order for each customer

from sqlalchemy import text

query = text("""
    SELECT o.customer_name, o.order_id, o.product
    FROM orders_no_id o
    INNER JOIN (
        SELECT customer_name, MIN(order_date) AS first_order_date
        FROM orders_no_id
        GROUP BY customer_name
    ) first_orders
    ON o.customer_name = first_orders.customer_name
    AND o.order_date = first_orders.first_order_date
    ORDER BY o.customer_name;
""")

with engine.connect() as conn:
    result = conn.execute(query)

    print("First order of each customer")
    for row in result:
        customer_name = row[0]
        order_id = row[1]
        product = row[2]
        print(f"Customer: {customer_name}, Order ID: {order_id}, Product: {product}")
```
